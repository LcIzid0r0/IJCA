IMPORTAÇÃO ARQUIVO PARQUETS

```python
file_path = '/content/drive/MyDrive/export_sd_chamados_historico_part-00000-tid-4316804262685302617-660de38f-1493-4f46-b3a8-4cc819403a2f-30-1-c000.snappy.parquet'
df = pd.read_parquet(file_path)
```

PADRONIZAÇÃO null para Não Informada

```python
df.description = df.description.replace({'null': 'Não Informada'})
```

PADRONIZAÇÃO PARA FORMATO CORRETO

```python
df['time.value'] = df['time.value'] .astype(int)
df['time.value'] = pd.to_datetime(df['time.value'], unit='ms', utc=True).dt.tz_convert('America/Sao_Paulo')
```

DROP DE COLUNAS

```python
df.drop('time.display_value', axis=1, inplace=True)
```

```python
df.drop('by.first_name', axis=1, inplace=True)
```

```python
df.drop('by.photo_url', axis=1, inplace=True)
```

```python
df.drop('by.last_name', axis=1, inplace=True)
```
